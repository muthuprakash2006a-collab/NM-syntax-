```python
from transformers import pipeline

model_pipe = pipeline(
    "text-generation",
    model="ibm-granite/granite-3.3-2b-instruct"
)
```

```python
import gradio as gr

def chat(user_message):
    response = model_pipe(user_message, max_new_tokens=200)
    return response[0]["generated_text"]

demo = gr.Interface(
    fn=chat,
    inputs="text",
    outputs="text",
    title="Syntax Surgeon Chatbot"
)

demo.launch(share=True)
```
